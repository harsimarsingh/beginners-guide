### Resnet as encoder to the Dynamic Unet using FastAI Library ###

```python
import fastai
from fastai.vision.models.unet import *
from fastai.callbacks import *
from fastai.utils.mem import *
```

```python
# class DynamicUnet(SequentialEx):
#     "Create a U-Net from a given architecture."
#     def __init__(self, encoder:nn.Module, n_classes:int, img_size:Tuple[int,int]=(256,256), blur:bool=False, blur_final=True, self_attention:bool=False,
#                  y_range:Optional[Tuple[float,float]]=None,
#                  last_cross:bool=True, bottle:bool=False, **kwargs):
```

```python
arch = torchvision.models.resnet34(pretrained=False)
```

```python
li = list(arch.children())
```

```python
arch = nn.Sequential(*li[:8])
```

```python
arch
```

```python
summary(arch.to('cuda'),input_size=(3, 320, 320))
```

```python
model = DynamicUnet(encoder=arch.to('cpu'),n_classes = 1,img_size = (320,320))
```

```python
summary(model.to('cuda'),input_size=(3, 320, 320))
```
